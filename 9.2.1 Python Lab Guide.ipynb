 ### 2.1.1 Deep Copy and Shallow Copy

 The `copy` module in Python is used to implement deep copy.

```python
import copy

original_list = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow_copied = copy.copy(original_list)
# Deep copy
deep_copied = copy.deepcopy(original_list)

# Modifying original list affects shallow copy but not deep copy
original_list[0][0] = 'X'

print("Original:", original_list)
print("Shallow Copy:", shallow_copied)
print("Deep Copy:", deep_copied)

```

 ### 2.1.2 if Statement

 ### 2.1.3 Loop Statement

 **Step 1: for loop**

 Use the `for` loop to generate a multiplication table.

```python
for i in range(1, 10):
    row = ""
    for j in range(1, i + 1):
        row += f"{j}x{i}={i*j}\t"
    print(row)
```

 **Step 2: while loop**

 When the condition is met, the statement block is executed cyclically. To end the loop, use `break` or `continue`.

```python
count = 0
while count < 5:
    print("Count is", count)
    count += 1
    if count == 3:
        print("Breaking loop early!")
        break

```

 # 3 Function and Object-oriented Experiment

 ## 3.1 Procedure

 ### 3.1.1 Function Customization

 **Step 1: Define a function.**

 Define a function to return a sequence. Each number in the sequence is the sum of the former two numbers (Fibonacci sequence).

```python
def fibonacci(n):
    """Generate Fibonacci sequence up to n elements."""
    seq = [0, 1]
    if n <= 2:
        return seq[:n]
    for _ in range(2, n):
        seq.append(seq[-1] + seq[-2])
    return seq

print("Fibonacci sequence (10 elements):", fibonacci(10))
```

 **Step 2: Generate parameters.**

 The function can be customized to generate parameters transferred in different ways.

```python
# Function with default arguments and variable length arguments
def custom_func(required_arg, default_arg="Default", *args, **kwargs):
    print("Required:", required_arg)
    print("Default Config:", default_arg)
    print("Extra Positional (*args):", args)
    print("Keyword Args (**kwargs):", kwargs)

custom_func("AI", "Override", 1, 2, 3, key="value")

```

 ### 3.1.2 Object Orientation

 **Step 1: Create and use classes.**

 Create the `Dog` class. Each instance created based on the `Dog` class stores the name and age. We will give each dog the `sit()` and `roll_over()` abilities.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sit(self):
        print(f"{self.name} is now sitting.")

    def roll_over(self):
        print(f"{self.name} rolled over!")

my_dog = Dog("Buddy", 4)
my_dog.sit()
my_dog.roll_over()
```

 **Step 2: Access attributes.**

```python
print(f"My dog's name is {my_dog.name} and he is {my_dog.age} years old.")
```

 **Step 3: Inherit a class.**

 One of the main benefits of object-oriented programming is code reuse, which is achieved through inheritance. Inheritance can be understood as the type and subtype relationships between classes.

```python
class GoldenRetriever(Dog):
    def fetch(self):
        print(f"{self.name} is fetching the ball!")

my_retriever = GoldenRetriever("Max", 2)
my_retriever.sit()
my_retriever.fetch()
```

 **Step 4: Class attributes and methods.**

```python
class DogWithSpecies(Dog):
    species = "Canis familiaris" # Class attribute

    @classmethod
    def print_species(cls):
        print(f"Species is {cls.species}")

DogWithSpecies.print_species()
dog_obj = DogWithSpecies("Rex", 3)
print(dog_obj.species)

```
