 # 1 I/O Operation Experiment

 ## 1.1 Procedure

 ### 1.1.1 I/O Operations

 **Step 1: Write data to a file.**

```python
filename = "experiment.txt"
content = "python operation experiment\n"
# Using standard open without context manager (as Step 3 covers context manager)
f = open(filename, "w")
f.write(content)
f.close()
print("Please enter the content to be written:\nthe content has been written")
```

 **Step 2: Read file data.**

```python
f = open(filename, "r")
read_content = f.read()
f.close()
print("Read from file:", read_content)
```

 **Step 3: Use the context manager to operate the file.**

```python
with open(filename, "r") as f:
    context_content = f.read()
print("Read using context manager:", context_content)
```

 **Step 4: Use the OS module to operate the file.**

```python
import os

print("Does file exist before deletion?", os.path.exists(filename))
os.remove(filename) # Cleaning up the file
print("Does file exist after deletion?", os.path.exists(filename))

```

 ## 2.3 Database Programming

 **Step 1: Install modules required for the database.**

 Run the pip command to install a third-party library in Python.

 `pip install pymysql`

 **Step 2: Connect to the database.**

```python
import pymysql

try:
    # Connect to the database.
    # localhost: local connection. You can also change it to the IP address of the database.
    # root: MySQL database account; mysql: database password.
    # my_database: name of the database to be connected.
    db = pymysql.connect(host="localhost", user="root", password="password", database="my_database")
    
    # Use the cursor() method to obtain the operation cursor.
    cursor = db.cursor()
    
    # Execute the SQL statement using the execute method.
    cursor.execute("SELECT VERSION()")
    
    # Use the fetchone() method to obtain a data record.
    data = cursor.fetchone()
    print("Database version : %s " % data)
except Exception as e:
    print("Skipping DB connection due to missing credentials or MySQL instance:", e)

```

 **Step 3: Use Python to operate the database.**

```python
try:
    # SQL statement for creating a data table.
    sql_create_table = """
    CREATE TABLE IF NOT EXISTS my_table (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL
    )
    """
    
    # Run the following commands:
    cursor.execute(sql_create_table)
    
except Exception as e:
    print("Database execution skipped:", e)
finally:
    try:
        # Close the database connection.
        db.close()
    except:
        pass

```

 **Step 4: Data operations**

```python
try:
    # Connect to the database again since it was closed above
    db = pymysql.connect(host="localhost", user="root", password="password", database="my_database")
    # Use the cursor() method to obtain the operation cursor.
    cursor = db.cursor()
    
    # Insert data
    # Execute the SQL INSERT statement.
    sql_insert = "INSERT INTO my_table (name) VALUES ('TestUser')"
    
    try:
        # Execute the SQL statement.
        cursor.execute(sql_insert)
        # Submit data to the database for execution.
        db.commit()
        print("Data inserted successfully")
    except:
        # Rollback upon an error.
        db.rollback()
        
    # View the result in the database.
    cursor.execute("SELECT * FROM my_table")
    results = cursor.fetchall()
    print("Data in my_table:", results)
    
    # Close the database connection.
    db.close()
except Exception as e:
    print("Data operation skipped:", e)
```
