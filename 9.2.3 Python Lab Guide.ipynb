 # 1 Regular Expression Experiment

 ## 1.1 Introduction

 ### 1.1.1 Regular Expressions



 **Step 1: `re.match` function**

 The `re.match` function attempts to match a pattern from the start position of the string. If the match is not successful, `match()` returns none.

```python
import re

text = "Huawei AI Training"
# match only searches from the beginning
match_obj = re.match(r'Huawei', text)
if match_obj:
    print("Match found:", match_obj.group())
else:
    print("No match!")

match_obj_fail = re.match(r'AI', text)
print("Match AI from start?", "Found" if match_obj_fail else "Not Found")
```

 **Step 2: `re.search` method**

 The `re.search` method scans the entire string and returns the first successful match.

```python
search_obj = re.search(r'AI', text)
if search_obj:
    print("Search found:", search_obj.group(), "at span:", search_obj.span())
else:
    print("No match!")
```

 **Step 3: Retrieval and replacement**

 The `re` module of Python provides `re.sub` function to replace the matching items in the string.

```python
text_with_numbers = "Phone number is 123-456-7890."
# Replace all digits with X
replaced_text = re.sub(r'\d', 'X', text_with_numbers)
print("Replaced text:", replaced_text)
```

 **Step 4: `re.compile` function**

 The `compile` function is used to compile regular expressions and generate a regular expression object.

```python
pattern = re.compile(r'\d+')
matches = pattern.findall("There are 10 apples and 5 oranges.")
print("Compiled sequence found numbers:", matches)
```

 **Step 5: `re.split()`**

 The `split` method splits a string based on the matched substring and returns a list.

```python
mixed_split = re.split(r'\W+', "Huawei,AI;Training.Python")
print("Split result:", mixed_split)

```

 ### 2.1.1 Iterator

 **Step 1: Determine an iterable object.**

```python
from collections.abc import Iterable

print("Is list iterable?", isinstance([1, 2, 3], Iterable))
print("Is integer iterable?", isinstance(123, Iterable))
```

 **Step 2: Obtain the iterator.**

 The `iter()` function is used to obtain the iterators of the iterable objects. You can use the `next()` function continuously.

```python
my_list = [10, 20, 30]
my_iter = iter(my_list)
print("1st yield:", next(my_iter))
print("2nd yield:", next(my_iter))
print("3rd yield:", next(my_iter))
```

 ### 2.1.2 Generators

 **Step 1: Create a generator through list derivation.**

```python
# Note: Generator expression uses (), list comprehension uses []
gen = (x * x for x in range(3))
print("Generator type:", type(gen))
for val in gen:
    print("Gen val:", val)
```

 **Step 2: Create a generator through a yield keyword.**

 Use the yield keyword to create a generator to generate the Fibonacci sequence.

```python
def fibonacci_gen(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

fib_g = fibonacci_gen(10)
for num in fib_g:
    print("Fibonacci generated:", num)
```

 **Step 3: Use the send method.**

 The `send` method can be used to wake up the generator and transfer data to the breakpoint.

```python
def echo_generator():
    while True:
        received = yield
        print(f"I received: {received}")

echo_g = echo_generator()
next(echo_g) # Prime the generator
echo_g.send("Hello")
echo_g.send("World!")
```

 ### 2.1.3 Decorators

 **Step 1 & 2: Construct a decorator and Use the decoration function.**

 Create a simple decorator that calculates the runtime of a function.

```python
import time

def calculate_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function {func.__name__} took {end_time - start_time:.6f} seconds to run.")
        return result
    return wrapper

@calculate_time
def slow_function(delay_seconds):
    time.sleep(delay_seconds)
    return "Done"

slow_function(0.5)
```
